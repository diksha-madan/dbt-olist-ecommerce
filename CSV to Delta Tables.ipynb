```python
import os

print(
  os.listdir(
    '/Volumes/olist/olist_raw/olist_csv'
  )
)
```

```python
all_files = os.listdir(
    '/Volumes/olist/olist_raw/olist_csv/'
  )

for i in all_files:
    file_name = i.split(".")[0]
    print(file_name)
    table_name = f"olist.olist_raw.{file_name}"

    full_path = f"/Volumes/olist/olist_raw/olist_csv/{i}"

    spark.sql(f"""CREATE TABLE if NOT EXISTS {table_name};""")

    spark.sql(f"""COPY INTO {table_name}
    FROM '{full_path}'
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true', 'inferSchema' = 'true')
    COPY_OPTIONS ('mergeSchema' = 'true'); """)
```
